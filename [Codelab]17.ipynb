11. Quantidade de estados por região?

```python
print(f"Estados do nordeste: {df_nordeste['Unidade federativa'].count()}")
print(f"Estados do norte: {df_norte['Unidade federativa'].count()}")
print(f"Estados do sul: {df_sul['Unidade federativa'].count()}")
print(f"Estados do sudeste: {df_sudeste['Unidade federativa'].count()}")
print(f"Estados do Centro-Oeste: {df_centro_oeste['Unidade federativa'].count()}")
```

12. Polulação total de cada região?

```python
print(f"População do sul: {df_sul['População'].sum()}")
print(f"População do sudeste: {df_sudeste['População'].sum()}")
print(f"População do nordeste: {df_nordeste['População'].sum()}")
print(f"População do centro-oeste: {df_centro_oeste['População'].sum()}")
print(f"População do norte: {df_norte['População'].sum()}")
```

14. Qual a regiçao com a menor população?

```python
regioes = df_estados['Região'].unique()
populacoes_regional = []

for i in regioes:
  filtro = df_estados['Região'] == i
  populacoes_regional.append(df_estados[filtro]['População'].sum())

print(f"A região mais populosa é a {regioes[populacoes_regional.index(max(populacoes_regional))]}")
print(f"A região menos populosa é a {regioes[populacoes_regional.index(min(populacoes_regional))]}")
```

15. Qual a população média de cada região?

```python
regioes = df_estados['Região'].unique()

for i in regioes:
  filtro = df_estados['Região'] == i
  print(f"A população média do {i} é {df_estados[filtro]['População'].mean()}")
```

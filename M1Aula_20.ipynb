```python
df['data_nasc']=pd.to_datetime(df['data_nasc'])
df['dia'] = df['data_nasc'].dt.day
df['mes'] = df['data_nasc'].dt.month
df['ano'] = df['data_nasc'].dt.year
df = df.drop(columns = ['nova_coluna'])
```

```python
def signos(x):
  if ((x['mes'] == 3) & (x['dia'] >=21)) | ((x['mes'] == 4) & (x['dia'] <= 20)):
    return 'Áries'
  elif ((x['mes'] == 4 ) & (x['dia'] >=21)) | ((x['mes'] == 5) & (x['dia'] <= 20)):
    return 'Touro'
  elif ((x['mes'] == 5 ) & (x['dia'] >=21)) | ((x['mes'] == 6) & (x['dia'] <= 20)):
    return 'Gêmeos'
  elif ((x['mes'] == 6 ) & (x['dia'] >=21)) | ((x['mes'] == 7) & (x['dia'] <= 22)):
    return 'Câncer'
  elif ((x['mes'] == 7 ) & (x['dia'] >=23)) | ((x['mes'] == 8) & (x['dia'] <= 22)):
    return 'Leão'
  elif ((x['mes'] == 8 ) & (x['dia'] >=23)) | ((x['mes'] == 9) & (x['dia'] <= 22)):
    return 'Virgem'
  elif ((x['mes'] == 9 ) & (x['dia'] >=23)) | ((x['mes'] == 10) & (x['dia'] <= 22)):
    return 'Libra'
  elif ((x['mes'] == 10 ) & (x['dia'] >=23)) | ((x['mes'] == 11) & (x['dia'] <= 21)):
    return 'Escorpião'
  elif ((x['mes'] == 11 ) & (x['dia'] >=22)) | ((x['mes'] == 12) & (x['dia'] <= 21)):
    return 'Sagitário'
  elif ((x['mes'] == 12 ) & (x['dia'] >=22)) | ((x['mes'] == 1) & (x['dia'] <= 19)):
    return 'Capricórnio'
  elif ((x['mes'] == 1 ) & (x['dia'] >=20)) | ((x['mes'] == 2) & (x['dia'] <= 18)):
    return 'Aquário'
  elif ((x['mes'] == 2 ) & (x['dia'] >=19)) | ((x['mes'] == 3) & (x['dia'] <= 20)):
    return 'Peixes'
```

```python
df.apply(signos, axis = 1)
df['Signos'] = df.apply(signos, axis = 1)
```

# EXERCICIO

Sortear 3 musicas do arquivo 'Musicas.csv'

Classificar as musicas restantes em grupo 1, grupo 2 e grupo 3 a partir das distancias entre as musicas de referencia (sorteadas) e os dados restantes

Plotar e separar por cor

```python
import math
import plotly.express as pt
```

```python
def dist_euclidiana(d1,d2,e1,e2): # Função de distancia, apenas para facilitar
  return math.sqrt(((d2-d1)**2) + ((e2-e1)**2))
```

```python
music = pd.read_csv('Musicas.csv')
```

```python
grupo = music.sample(3).reset_index() # Amostra 3 valores do conjunto de dados para serem a referencia

# Guarda os valores de dancabilidade e energia das musicas que irão definir cada grupo

grupo1 = [grupo['danceability'][0],grupo['energy'][0]]
grupo2 = [grupo['danceability'][1],grupo['energy'][1]]
grupo3 = [grupo['danceability'][2],grupo['energy'][2]]
```

```python
vetor_distancias = []
vetor_classe = []

# Calcula a distancia da musica atual em relação as musicas de referencia
# guarda os 3 valores num vetor de distancia, 
# encontra o valor minimo do vetor de distancia e retorna seu indice
# que será correspondente ao grupo, faz um vetor de grupos

for i in range(music.shape[0]):
  dist1 = dist_euclidiana(grupo1[0],music['danceability'][i],grupo1[1],music['energy'][i])
  dist2 = dist_euclidiana(grupo2[0],music['danceability'][i],grupo2[1],music['energy'][i])
  dist3 = dist_euclidiana(grupo3[0],music['danceability'][i],grupo3[1],music['energy'][i])
  vetor_distancias.append(dist1)
  vetor_distancias.append(dist2)
  vetor_distancias.append(dist3)
  menor_dist = vetor_distancias.index(min(vetor_distancias))
  vetor_distancias = []
  vetor_classe.append(f"Grupo {menor_dist+1}")
  
music['Grupo'] = vetor_classe # Vetor de grupos é adicionado ao dataset na coluna 'Grupos'
```

```python
# As figuras são plotadas num grafico de danceability x energy e rotulados (com cores) pela classificação de grupo

fig = pt.scatter(music, x="danceability", y="energy", color="Grupo")
fig.show()
```
